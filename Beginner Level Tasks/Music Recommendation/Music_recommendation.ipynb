```python
members = pd.read_csv('/drive/My Drive/Music recommendation datasets/members.csv')
songs =  pd.read_csv('/drive/My Drive/Music recommendation datasets/songs.csv')
train =  pd.read_csv('/drive/My Drive/Music recommendation datasets/train.csv')
```

```python
train = pd.merge(train, songs, on="song_id", how="left")
```

```python
train.shape
```

```python
train.head()
```

```python
train = pd.merge(train, members, on="msno", how="left")
```

```python
for i in train.select_dtypes(include=['object']).columns:
    train[i][train[i].isnull()] = 'unknown'
train = train.fillna(value=0)
```

```python
train.isnull().sum()
```

```python
train.head()
```

```python
for col in train.select_dtypes(include=['object']).columns:
    train[col] = train[col].astype('category')
for i in train.select_dtypes(include=['category']).columns:
    train[i] = train[i].cat.codes
```

```python
X_train = train.drop(['target'], axis = 1).values
y_train = train['target'].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train,test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10)
```

```python
rfc.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
rfc_pred = rfc.predict(X_test)
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```
